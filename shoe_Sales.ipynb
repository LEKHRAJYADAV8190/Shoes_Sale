```python
import matplotlib.pyplot as plt

# Create the figure
plt.figure(figsize=(10, 6))

# Plot sales for df_a
plt.plot(df_a['date'], df_a['sold_qty'], marker='o', label='Brand A Sales')

# Plot sales for df_nike
plt.plot(df_nike['date'], df_nike['sold_qty'], marker='s', label='Nike Sales')

# Add titles and labels
plt.title('Sales Analysis of Brand A and Nike')
plt.xlabel('Date')
plt.ylabel('Sold Quantity')
plt.legend()

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.tight_layout()  # Ensures everything fits without overlap
plt.show()

```
